```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(10)
```

**Exercício 1:**

image.png

```python
m = 1000
x = np.random.rand(m,1)
w = np.random.randn(m,1)
y = 1.2 + 2.3*x
y_noisy = 1.2 + 2.3*x + 10*w
```

```python
plt.scatter(x,y_noisy, label='funcao ruidosa')
plt.scatter(x,y,label='funcao exata')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

**Exercício 2:**

image.png

```python
m = 10000
x = np.random.rand(m,1)
y = np.random.rand(m,1)
z = x + y

plt.hist(z,bins=11)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

image.png

```python
x1 = np.arange(-10,11,0.25)
x2 = np.arange(-10,11,0.25)
(x1,x2) = np.meshgrid(x1,x2)
y = x1**2 + x2**2

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(x1,x2,y)
```
