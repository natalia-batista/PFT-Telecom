**Seção 11 — Raise**

**Exercício 18**

Crie uma função chamada `validar_senha(senha)` que deve verificar:

* A senha deve ter pelo menos 8 caracteres
* Deve conter pelo menos um número
* Deve conter pelo menos uma letra maiúscula

Se qualquer uma dessas regras não for cumprida, a função deve lançar uma exceção personalizada chamada:

```
class SenhaInvalidaError(Exception):
    pass
```



A função deve retornar `"Senha válida!"` se tudo estiver correto.

````python
class SenhaInvalidaError(Exception):
    pass

flag_upper = False
flag_number = False

def validar_senha(senha):
  if len(senha)<8:
    raise SenhaInvalidaError("Senha inválida")

  for i in range(len(senha)):
    if senha[i].isupper():
      flag_upper = True
    elif senha[i].isdigit():
      flag_number = True

  if flag_upper and flag_number:
    print('Senha válida')
  else:
    raise SenhaInvalidaError("Senha invalida")



validar_senha('Natalia233')
````

**Seção 12 — Desafio**

**Exercício 19**

Crie um programa que:

1. Peça nome e idade de 5 pessoas

2. Armazene em uma lista de dicionários

3. Informe quem é a pessoa mais velha

````python
lista = []
for i in range(5):
  nome = input('Nome: ')
  idade = int(input('Idade: '))
  dicionario = {'Nome':nome, 'Idade':idade}
  lista.append(dicionario)

print(lista)
````

````python
id_max = 0
for i in range(len(lista)):
  dicionario = lista[i]
  if dicionario['Idade'] > id_max:
    pessoa_mais_velha = dicionario['Nome']
    id_max = dicionario['Idade']

print(pessoa_mais_velha)
````
